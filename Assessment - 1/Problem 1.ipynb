## TASK 1 : To Create Perceptron Model Using Basic Python.

```python
train_data = [[11,12,13],
        [16,34,25],
        [36,36,22],
        [22,22,11]]
test_data=[0,0,1,1]
```

```python
print(train_data)
```

```python
print(test_data)
```

```python
epochs = 3
l_rate = 1
w1 = 0  
w2 = 0
w3 =  0
unit_bias = 1
bias = 0
y_predict = 0
```

```python
for epc in range(0,epochs):
    print("\nEpoch-"+str(epc+1))
    print("----------------------------------------------------------------")
    print("----------------------------------------------------------------")
    for i,x in zip(train_data,test_data):
        input1 = i[0]
        input2 = i[1]
        input3 = i[2]
        y_in = bias + (input1 * w1)+(input2*w2)+(input3*w3)
        if y_in > 0:
            y=1
        elif y_in == 0:
            y=0
        elif y_in <0 :
            y=-1
        else:
            print("something went wrong")
        print("input1 is : - ",input1)
        print("input2 is :- ",input2)
        print("input3 is :-",input3)
        print("Unit_bias is :-",1)
        print("Target is :- ",x)
        if y!=x:
            w1 =( l_rate * x * input1) + w1
            w2 =( l_rate * x * input2) + w2
            w3 =( l_rate * x * input3) + w3
            w1_old,w2_old,w3_old,bias_old = w1,w2,w3,bias
            bias = bias + (l_rate * x)
            print("Y_in is:-",y_in)
            print("Y is :-",y)
            print("Target :-",x)
            print("Change Weight-1 :-", str(l_rate * x * input1))
            print("Change Weight-2 :-", str(l_rate * x * input2))
            print("Change Weight-3 :- ", str(l_rate * x * input3))
            print("Change bias :- ",l_rate * x)
            print("New w1 :- ",w1)
            print("New w2 :- ",w2)
            print("New w3 :- ",w3)
            print("New Bias ",str(bias_old) +" + "+ str(l_rate * x))
        else:
            w1 = w1
            w2 = w2
            w3 = w3
            y_predict = y_predict + 1
            print("New w1 :- ",'old w1 = ',w1)
            print("New w2 :- ",'old w2 = ',w2)
            print("New w3 :- ",'old w3 = ',w3)
            print("New Bias :- ",'bias =',bias)
        print("\n\n")
```

```python
print("Accuracy Score is :- ",y_predict,"/",(len(test_data))*epochs,"=",(y_predict/(len(test_data)*epochs)))
print((y_predict/(len(test_data)*epochs))*100,"%")
```
